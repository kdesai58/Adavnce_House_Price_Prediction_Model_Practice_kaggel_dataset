```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('house-prices-advanced-regression-techniques/X_train.csv')
```

```python
df
```

```python
y = df['SalePrice']
y
```

```python
X = df.drop(columns='SalePrice')
X.head(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
y_train
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

### LET'S TRY WITH COST FUNCTION

```python
sgdr = SGDRegressor(max_iter=1000)
sgdr.fit(X_train, y_train)
print(sgdr)
print(f"number of iterations completed: {sgdr.n_iter_}, number of weight updates: {sgdr.t_}")
```

```python
y_pred_sgd = sgdr.predict(X_train)
```

```python
print(y_test.shape)
print(y_pred_sgd.shape)
```

```python
y_test
```

```python
y_pred_sgd
```

```python
mse = mean_squared_error(y_train, y_pred_sgd)
r2 = r2_score(y_train, y_pred_sgd)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
